### Import Libraries

```python
import pandas as pd
import numpy as np
from numpy import nan
import matplotlib.pyplot as plt


from numpy import array
from keras.models import Sequential
from keras.layers import Dense

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, explained_variance_score
from sklearn.linear_model import LinearRegression


from numpy import array

import warnings
warnings.filterwarnings('ignore')
```

### Load Data

```python
# load all data
dataset = pd.read_csv('household_power_consumption.txt', sep=';', header=0, low_memory=False, infer_datetime_format=True, parse_dates={'datetime':[0,1]}, index_col=['datetime'])
# summarize
dataset.head()
```

Only 'Global_active_power' is considered for prediction.

```python
dataset= dataset['Global_active_power']
```

Store the indices of the missing values.

```python
nan_value= np.where(np.asarray(dataset)== '?')
```

Delete those entries.

```python
dataset= np.delete(np.asarray(dataset), nan_value, 0)
```

### Data Preparation

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the sequence
        if end_ix > len(sequence)-1:
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.asarray(X), np.asarray(y)


# choose a number of time steps
n_steps = 60
# split into samples
dataset, labels = split_sequence(dataset, n_steps)


r2_scores= []
mean_squared_errors= []
mean_absolute_errors= []
explained_variance_scores= []
classifers= ['MLP', 'Linear Reg']
```

### Split Data into Train and Validation

```python
def split_data(dataset, labels):
    length= int(0.2*len(dataset))
    train_data = dataset[length:]
    valid_data = dataset[:length]
    train_labels= labels[length:]
    valid_labels= labels[:length]

    return train_data, train_labels, valid_data, valid_labels


train_data, train_labels, valid_data, valid_labels= split_data(dataset, labels)
```

## Multilayer Perceptrons (MLP)

A multilayer perceptron (MLP) is a deep, artificial neural network. It is composed of more than one perceptron. They are composed of an input layer to receive the signal, an output layer that makes a decision or prediction about the input, and in between those two, an arbitrary number of hidden layers that are the true computational engine of the MLP. MLPs with one hidden layer are capable of approximating any continuous function.

Multilayer perceptrons are often applied to supervised learning problems3: they train on a set of input-output pairs and learn to model the correlation (or dependencies) between those inputs and outputs. Training involves adjusting the parameters, or the weights and biases, of the model in order to minimize error. Backpropagation is used to make those weigh and bias adjustments relative to the error, and the error itself can be measured in a variety of ways, including by root mean squared error (RMSE).

## Univariate MLP Models

Multilayer Perceptrons, or MLPs for short, can be used to model univariate time series forecasting problems.

Univariate time series are a dataset comprised of a single series of observations with a temporal ordering and a model is required to learn from the series of past observations to predict the next value in the sequence.

# Learning using MLP.

```python
# define model
model = Sequential()
model.add(Dense(100, activation='relu', input_dim=n_steps))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# fit model
model.fit(train_data, train_labels, epochs=1, verbose=0)

# predict on validation data
predictions= []
for i in range(len(valid_data)):
    predict_power= model.predict(np.asarray(valid_data[i]).reshape(1, n_steps), verbose=0)
    predictions.append(predict_power[0][0])
    
# Display metrics
metric= r2_score(valid_labels.astype(np.float32), predictions)
r2_scores.append(metric)
print ("R2 score on Validation Data using MLP= ", metric)
metric= mean_squared_error(valid_labels.astype(np.float32), predictions)
mean_squared_errors.append(metric)
print("Mean Squared Error on Validation Data using MLP= ", metric)
metric= mean_absolute_error(valid_labels.astype(np.float32), predictions)
mean_absolute_errors.append(metric)
print("Mean Absolute Error on Validation Data using MLP= ", metric)
metric= explained_variance_score(valid_labels.astype(np.float32), predictions)
explained_variance_scores.append(metric)
print("Explained Variance Score on Validation Data using MLP= ", metric)
```

## Linear Regression


1 Its a supervised machine learning algorithm which learns from given x dependent variable and Y as quantifiable variable and predicts New Y from given new X.

2 It works in a way such that it finds the best fit line from the given data.

Clearly stated, the goal of linear regression is to fit a line to a set of points.

### Gradient Descent

Gradient descent is an optimization algorithm used to find the values of parameters (coefficients) of a function (f) that minimizes a cost function (cost).

### Learning Rate

Pick a value for the learning rate . The learning rate determines how big the step would be on each iteration. If  is very small, it would take long time to converge and become computationally expensive. If  is large, it may fail to converge and overshoot the minimum.

### Cost Function & Gradient

# Learning using Linear Regression.

```python
regressor = LinearRegression()  
regressor.fit(train_data, train_labels) #training the algorithm

# predict on validation data
predictions = regressor.predict(valid_data)

# Display metrics
metric= r2_score(valid_labels.astype(np.float32), predictions)
r2_scores.append(metric)
print ("R2 score on Validation Data using Linear Regression= ", metric)
metric= mean_squared_error(valid_labels.astype(np.float32), predictions)
mean_squared_errors.append(metric)
print("Mean Squared Error on Validation Data using Linear Regression= ", metric)
metric= mean_absolute_error(valid_labels.astype(np.float32), predictions)
mean_absolute_errors.append(metric)
print("Mean Absolute Error on Validation Data using Linear Regression= ", metric)
metric= explained_variance_score(valid_labels.astype(np.float32), predictions)
explained_variance_scores.append(metric)
print("Explained Variance Score on Validation Data using Linear Regression= ", metric)
```

## Comparision of MLP & Linear Regression (as per the studies)

Neural networks are used increasingly as statistical models. The performance of multilayer perceptron (MLP) and that of linear regression (LR) were compared, with regard to the quality of prediction and estimation and the robustness to deviations from underlying assumptions of normality, homoscedasticity and independence of errors. Taking into account those deviations, five designs were constructed, and, for each of them, 3000 data were simulated. The comparison between connectionist and linear models was achieved by graphic means including prediction intervals, as well as by classical criteria including goodness-of-fit and relative errors. The empirical distribution of estimations and the stability of MLP and LR were studied by re-sampling methods. MLP and linear regression had comparable performance and robustness. Despite the flexibility of connectionist models, their predictions were stable. The empirical variances of weight estimations result from the distributed representation of the information among the processing elements. This emphasizes the major role of variances of weight estimations in the interpretation of neural networks. This needs, however, to be confirmed by further studies. Therefore MLP could be useful statistical models, as long as convergence conditions are respected.
